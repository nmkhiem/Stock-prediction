```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
import GPy
from IPython.display import display
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('C:/Users/84909/Finpros/datasets/FPT.csv', index_col ="Date/Time")
df.index = pd.to_datetime(df.index)
df = df.resample('1D').mean().dropna()
df['Day'] = df.index.day
df['Year'], df['Month'] = df.index.year, df.index.month
df.head()
```

```python
df_train = df[:400]
df_train.shape
```

```python
X_train = df_train.iloc[:, 5:8]
y_train = df_train['Close'].values.reshape(-1, 1)
```

```python
kernel1 = GPy.kern.Matern32(input_dim=3)
kernel2 = GPy.kern.Poly(input_dim=3)
kernel3 = GPy.kern.PeriodicExponential(input_dim = 1)
kernel = kernel1 * kernel2
model = GPy.models.GPRegression(X_train, y_train, kernel)
```

```python
model.optimize()
```

```python
df_pred = df[400:]
df_pred.shape
```

```python
X_test = df_pred.iloc[:, 5:8]
y_test = df_pred['Close'].values.reshape(-1, 1)
```

```python
forecast, cov = model.predict(np.array(X_test))
```

```python
sigma = np.sqrt(cov)
```

```python
lower = forecast - 2*sigma
upper = forecast + 2*sigma
```

```python
plt.plot(forecast)
```

```python
plt.plot(df.Close[-200:])
plt.plot(df.Close.index[-len(y_test):], forecast)
plt.fill_between(df.Close.index[-len(y_test):], 
                 lower.squeeze(), 
                 upper.squeeze(), color='lightblue', alpha=0.5, label='95% Confidence Interval')
plt.show()
```

```python
from sklearn import metrics
```

```python
binary_true = np.where(pd.DataFrame(forecast).diff() > 0, 1, 0)
binary_pred = np.where(pd.DataFrame(y_test).diff() > 0, 1, 0)
```

```python
confusion_matrix = metrics.confusion_matrix(binary_true, binary_pred)
```

```python
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [0, 1])
```

```python
cm_display.plot()
plt.show()
```
