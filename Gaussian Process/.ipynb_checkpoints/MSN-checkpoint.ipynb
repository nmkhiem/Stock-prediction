```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler
import matplotlib.dates as mdates
import numpy as np
import os
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import ConstantKernel, RBF, ExpSineSquared, RationalQuadratic
import warnings
warnings.filterwarnings('ignore')
companies = []
```

```python
df = pd.read_csv('C:/Users/84909/Finpros/datasets/FPT.csv', index_col ="Date/Time")
```

```python
df = df.iloc[:, 4:5]
df = df.dropna()
df.index = pd.to_datetime(df.index)
df['Day'] = df.index.day
df['Year'], df['Month'] = df.index.year, df.index.month
```

```python
df = df.resample('12H').mean().dropna()
```

```python
df['log_returns'] = np.log(df.Close).diff()
df.dropna(inplace=True)
```

```python
df.head()
```

```python
index = np.array([])
for year in df['Year'].unique():
    close = df.Close[df['Year'] == year]
    index = np.append(index, range(0, len(close)))
```

```python
df['Index'] = index
df
```

```python
start_train = '2019-12-25'
end_train = '2020-10-31'
df_train = df[(df.index >= start_train) & (df.index <= end_train)].iloc[:, :]
df_train.head()
```

```python
X_train = np.array(df_train[['Year', 'Month', 'Day', 'Index']])
y_train = df_train.iloc[:, 4].values
len(X_train)
```

```python
start_predict = '2020-10-31'
end_predict = '2020-12-22'
df_predict = df[(df.index >= start_predict) & (df.index <= end_predict)]
df_predict.head()
```

```python
X_predict = np.array(df_predict[['Year', 'Month', 'Day', 'Index']])
y_test = df_predict.iloc[:, 4].values
len(X_predict)
```

```python
kernel = ConstantKernel(constant_value = 1.0) * RBF(length_scale = 1.) * ExpSineSquared() 
alpha = 1e-10
iterations = 30

gp = GaussianProcessRegressor(kernel = kernel, alpha = alpha, n_restarts_optimizer = iterations, normalize_y=False)
gp.fit(X_train, y_train)

print(gp.kernel_.get_params())
```

```python
y_pred, y_cov = gp.predict(X_predict, return_cov = True)

gp.score(X_predict, y_test)
```

```python
y_lower = y_pred - np.sqrt(np.diag(y_cov))
y_upper = y_pred + np.sqrt(np.diag(y_cov))
```

```python
plt.figure(figsize=(20, 5))
plt.plot(df_predict.index, y_test, label = 'test')
plt.plot(df_predict.index, y_pred, label = 'predict')
plt.fill_between(df_predict.index, y_lower, y_upper, alpha = 0.25, label = 'Variance')
plt.legend()
plt.show()
```

```python
scaler = StandardScaler()

y_test_scaled = scaler.fit_transform(y_test.reshape(-1, 1))
y_pred_scaled = scaler.fit_transform(y_pred.reshape(-1, 1))
```

```python
plt.plot(y_test_scaled)
plt.plot(y_pred_scaled)
plt.show()
```

```python
def log_returns_to_close(log_returns, initial_close_price):
    cumulative_log_returns = np.cumsum(log_returns)
    close_prices = initial_close_price * np.exp(cumulative_log_returns)
    return close_prices
```

```python
def confidence_interval_to_close(confidence_interval):
    lower_bound = np.exp(confidence_interval[0])
    upper_bound = np.exp(confidence_interval[1])
    return lower_bound, upper_bound
```

```python
y_close_pred = log_returns_to_close(y_pred, df.Close[-y_pred.shape[0]])
```

```python
y_pred_lower_close, y_pred_upper_close = confidence_interval_to_close([y_lower, y_upper])
```

```python
import seaborn as sns
plt.figure(figsize=(15, 5))
sns.set_style("ticks")
sns.lineplot(data = df[:], x = "Date/Time", y = 'Close')
sns.lineplot(data = df[-y_pred.shape[0]:], x = "Date/Time", y = 'Close')
sns.lineplot(data = df[-y_pred.shape[0]:], x = "Date/Time", y = y_close_pred)
plt.fill_between(df[-y_pred.shape[0]:].index, y_close_pred - y_pred_lower_close, y_close_pred + y_pred_upper_close, color='gray', alpha=0.3, label='Confidence Interval')
plt.title("The Stock Prices of FPT",size = 'x-large')
sns.despine()
plt.show()
```

## Volatility - Day

```python
df = pd.read_csv('C:/Users/84909/Finpros/datasets/FPT.csv', index_col ="Date/Time")
```

```python
df = df.iloc[:, 4:5]
df = df.dropna()
df.index = pd.to_datetime(df.index)
df['Day'] = df.index.day
df['Year'], df['Month'] = df.index.year, df.index.month
```

```python
df = df.resample('1D').mean().dropna()
```

```python
df.head()
```

```python
index = np.array([])
for year in df['Year'].unique():
    close = df.Close[df['Year'] == year]
    index = np.append(index, range(0, len(close)))
```

```python
df['Index'] = index
df['vol_5'] = df['Close'].rolling(5).std()
df.dropna(inplace = True)
df
```

```python
start_train = '2018-12-25'
end_train = '2019-12-31'
df_train = df[(df.index >= start_train) & (df.index <= end_train)].iloc[:, :]
df_train.head()
```

```python
X_train = np.array(df_train[['Year', 'Month', 'Day', 'Index']])
y_train = df_train.iloc[:, 5].values
len(X_train)
```

```python
start_predict = '2020-01-01'
end_predict = '2020-12-22'
df_predict = df[(df.index >= start_predict) & (df.index <= end_predict)]
df_predict.head()
```

```python
X_predict = np.array(df_predict[['Year', 'Month', 'Day', 'Index']])
y_test = df_predict.iloc[:, 5].values
len(X_predict)
```

```python
kernel = ConstantKernel(constant_value = 1.0) * RBF(length_scale = 1.) * ExpSineSquared() * RationalQuadratic()
alpha = 1e-10
iterations = 30

gp = GaussianProcessRegressor(kernel = kernel, alpha = alpha, n_restarts_optimizer = iterations, normalize_y=False)
gp.fit(X_train, y_train)

print(gp.kernel_.get_params())
```

```python
y_pred, y_cov = gp.predict(X_predict, return_cov = True)

gp.score(X_predict, y_test)
```

```python
y_lower = y_pred - np.sqrt(np.diag(y_cov))
y_upper = y_pred + np.sqrt(np.diag(y_cov))
```

```python
plt.figure(figsize=(20, 5))
plt.plot(df_predict.index, y_test, label = 'test')
plt.plot(df_predict.index, y_pred, label = 'predict')
plt.fill_between(df_predict.index, y_lower, y_upper, alpha = 0.25, label = 'Variance')
plt.legend()
plt.show()
```

```python
def std_to_close(std_values, initial_close_price):
    close_prices = []
    current_close = initial_close_price
    for std in std_values:
        current_close = current_close * (1 + std)
        close_prices.append(current_close)
    return close_prices
```

```python
close_from_std = std_to_close(y_pred, df.Close[-y_pred.shape[0]:].values)
```

```python
y_lower = std_to_close(y_lower, df.Close[-y_pred.shape[0]:].values)
y_upper = std_to_close(y_upper, df.Close[-y_pred.shape[0]:].values)
```

```python
bias = df[-y_pred.shape[0]:].Close[0] - close_from_std[0][0]
```
